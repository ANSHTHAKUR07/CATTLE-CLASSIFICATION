```python
# IMPORTANT: RUN THIS CELL IN ORDER TO IMPORT YOUR KAGGLE DATA SOURCES
import kagglehub

# Download latest version
path = kagglehub.dataset_download("lukex9442/indian-bovine-breeds")

print("Path to dataset files:", path)
print('Data source import complete.')


from pathlib import Path

root = Path(path)

for p in root.rglob("*"):
    if p.is_dir():
        subdirs = [d for d in p.iterdir() if d.is_dir()]
        if len(subdirs) > 5:  # more than 5 breeds => correct folder
            print("Found class root:", p)
            print("Subfolders:", [d.name for d in subdirs[:20]])
            break

# -----------------------------
# Full, corrected, end-to-end training + evaluation script
# Works with kagglehub.dataset_download('lukex9442/indian-bovine-breeds')
# -----------------------------

import os
import sys
from pathlib import Path

# --- locate dataset path returned by kagglehub ---
DATA_ROOT = Path(path) if 'path' in globals() else None
if DATA_ROOT is None or not DATA_ROOT.exists():
    raise RuntimeError(f"Dataset path from kagglehub not found: {DATA_ROOT}. Make sure the kagglehub download succeeded.")

# find the first directory inside DATA_ROOT that contains subfolders (classes) or many images
def find_image_root(root: Path):
    # If root itself looks like an image folder (contains class subdirs), return it
    children = [p for p in root.iterdir() if p.is_dir()]
    if not children:
        # maybe images directly in root
        return root
    # prefer a directory whose subdirectories contain many image files
    best = None
    best_count = 0
    for c in children:
        # count image files in child and its subdirs
        cnt = sum(1 for _ in c.rglob('*.jpg')) + sum(1 for _ in c.rglob('*.png'))
        if cnt > best_count:
            best_count = cnt
            best = c
    return best if best is not None else root

# OVERRIDE: Use the correct class root found earlier
DATA_DIR = "/root/.cache/kagglehub/datasets/lukex9442/indian-bovine-breeds/versions/5/Indian_bovine_breeds/Indian_bovine_breeds"

print("Using corrected DATA_DIR:", DATA_DIR)


# WORK_DIR should be writable
WORK_DIR = Path('./working')
WORK_DIR.mkdir(exist_ok=True)

# ----- imports (again, defensive) -----
import random, time, warnings
warnings.filterwarnings('ignore')

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset
from torchvision import datasets, transforms, models

from sklearn.metrics import (accuracy_score, precision_score, recall_score,
                             f1_score, confusion_matrix, classification_report)
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA

# reproducibility
SEED = 42
random.seed(SEED)
np.random.seed(SEED)
if 'torch' in globals():
    torch.manual_seed(SEED)
    if torch.cuda.is_available():
        torch.cuda.manual_seed_all(SEED)
        torch.backends.cudnn.deterministic = False
        torch.backends.cudnn.benchmark = True

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
PIN = device.type == 'cuda'
print('Using device:', device)

# --- transforms ---
train_transform = transforms.Compose([
    transforms.RandomResizedCrop(224, scale=(0.7, 1.0)),
    transforms.RandomHorizontalFlip(p=0.5),
    transforms.RandomVerticalFlip(p=0.2),
    transforms.RandomRotation(30),
    transforms.ColorJitter(brightness=0.3, contrast=0.3, saturation=0.3, hue=0.05),
    transforms.RandomGrayscale(p=0.05),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
    transforms.RandomErasing(p=0.2, scale=(0.02, 0.33))
])

valid_transform = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])

# --- dataset and split ---
print('Building ImageFolder from:', DATA_DIR)
base_dataset = datasets.ImageFolder(root=str(DATA_DIR), transform=valid_transform)
class_names = base_dataset.classes
num_classes = len(class_names)
print(f'Found {num_classes} classes')

from collections import defaultdict

def stratified_split(dataset, train_ratio=0.8, val_ratio=0.1, seed=SEED):
    rng = np.random.default_rng(seed)
    idxs_by_class = defaultdict(list)
    for idx, (_, lbl) in enumerate(dataset.samples):
        idxs_by_class[lbl].append(idx)
    train_idx, val_idx, test_idx = [], [], []
    for lbl, idxs in idxs_by_class.items():
        idxs = np.array(idxs)
        rng.shuffle(idxs)
        n = len(idxs)
        n_train = int(n * train_ratio)
        n_val = int(n * val_ratio)
        train_idx.extend(idxs[:n_train].tolist())
        val_idx.extend(idxs[n_train:n_train + n_val].tolist())
        test_idx.extend(idxs[n_train + n_val:].tolist())
    return train_idx, val_idx, test_idx

train_idx, val_idx, test_idx = stratified_split(base_dataset)

# Subset dataset with transform
class SubsetWithTransform(Dataset):
    def __init__(self, base_dataset, indices, transform):
        self.base = base_dataset
        self.indices = list(indices)
        self.transform = transform
        self.loader = base_dataset.loader
        self.samples = base_dataset.samples
        self.targets = [self.samples[i][1] for i in self.indices]

    def __len__(self):
        return len(self.indices)

    def __getitem__(self, idx):
        real_idx = self.indices[idx]
        path, target = self.samples[real_idx]
        img = self.loader(path)
        if self.transform:
            img = self.transform(img)
        return img, target

train_ds = SubsetWithTransform(base_dataset, train_idx, train_transform)
val_ds   = SubsetWithTransform(base_dataset, val_idx, valid_transform)
test_ds  = SubsetWithTransform(base_dataset, test_idx, valid_transform)

from sklearn.utils.class_weight import compute_class_weight

labels = [base_dataset.samples[i][1] for i in train_idx]

all_classes = np.arange(num_classes)

class_weights = compute_class_weight(
    class_weight='balanced',
    classes=all_classes,
    y=labels
)

class_weights = torch.tensor(class_weights, dtype=torch.float).to(device)

print('Sizes ->', len(train_ds), len(val_ds), len(test_ds))

batch_size = 32
num_workers = min(8, max(0, os.cpu_count() - 1))


train_loader = DataLoader(
    train_ds,
    batch_size=batch_size,
    shuffle=True,
    num_workers=num_workers,
    pin_memory=PIN,
    drop_last=True
)
val_loader = DataLoader(val_ds, batch_size=batch_size, shuffle=False,
                        num_workers=num_workers, pin_memory=PIN)
test_loader = DataLoader(test_ds, batch_size=batch_size, shuffle=False,
                         num_workers=num_workers, pin_memory=PIN)

# --- Model (EfficientNet B3 by default, fallback to ResNet50) ---
import torch
import torch.nn as nn
import torchvision.models as models

# -------- Attention Block (SE Block) --------
class SEBlock(nn.Module):
    def __init__(self, in_features, reduction=16):
        super(SEBlock, self).__init__()
        self.fc = nn.Sequential(
            nn.Linear(in_features, in_features // reduction),
            nn.ReLU(),
            nn.Linear(in_features // reduction, in_features),
            nn.Sigmoid()
        )

    def forward(self, x):
        weights = self.fc(x)
        return x * weights


# -------- HYBRID MODEL --------
class HybridCattleClassifier(nn.Module):
    def __init__(self, num_classes):
        super().__init__()

        self.effnet = models.efficientnet_b3(weights="IMAGENET1K_V1")
        eff_dim = self.effnet.classifier[1].in_features
        self.effnet.classifier = nn.Identity()

        self.resnet = models.resnet50(weights="IMAGENET1K_V2")
        res_dim = self.resnet.fc.in_features
        self.resnet.fc = nn.Identity()

        self.eff_proj = nn.Linear(eff_dim, 512)
        self.res_proj = nn.Linear(res_dim, 512)

        # ✅ FIXED: defined once, not inside forward
        self.fusion = nn.Linear(1024, 512)

        self.attention = SEBlock(512)

        self.classifier = nn.Sequential(
            nn.Linear(512, 256),
            nn.BatchNorm1d(256),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(256, num_classes)
        )

    def forward(self, x):
        f1 = self.effnet(x)
        f2 = self.resnet(x)

        f1 = self.eff_proj(f1)
        f2 = self.res_proj(f2)

        fused = torch.cat([f1, f2], dim=1)
        fused = self.fusion(fused)
        fused = self.attention(torch.relu(fused))

        return self.classifier(fused)

    @torch.no_grad()
    def get_features(self, x):
        f1 = self.effnet(x)
        f2 = self.resnet(x)

        f1 = self.eff_proj(f1)
        f2 = self.res_proj(f2)

        return (f1 + f2) / 2
# utility: denormalize for plotting
def denorm(img_tensor):
    img = img_tensor.detach().cpu().numpy().transpose(1,2,0)
    img = np.clip(img * np.array([0.229,0.224,0.225]) + np.array([0.485,0.456,0.406]), 0,1)
    return img

# 🔥 1. Define mixup FIRST (or anywhere above usage)
# mixup
def mixup(x, y, alpha=0.4):
    lam = np.random.beta(alpha, alpha)
    index = torch.randperm(x.size(0)).to(x.device)
    mixed_x = lam * x + (1 - lam) * x[index]
    y_a, y_b = y, y[index]
    return mixed_x, y_a, y_b, lam


# ✅ CORRECT POSITION
class FocalLoss(nn.Module):
    def __init__(self, gamma=2, weight=None):
        super().__init__()
        self.gamma = gamma
        self.weight = weight

    def forward(self, inputs, targets):
        ce_loss = nn.CrossEntropyLoss(
            weight=self.weight,
            reduction='none'
        )(inputs, targets)

        pt = torch.exp(-ce_loss)
        loss = ((1 - pt) ** self.gamma) * ce_loss
        return loss.mean()


def train_model(model, train_loader, val_loader, epochs=50, patience=13, lr=2e-4):

    model = model.to(device)

    # ✅ Correct loss
    criterion = FocalLoss(gamma=2, weight=class_weights)

    optimizer = optim.AdamW(model.parameters(), lr=lr, weight_decay=1e-2)

    scheduler = optim.lr_scheduler.CosineAnnealingWarmRestarts(
        optimizer, T_0=10, T_mult=2, eta_min=1e-6
    )

    best_val = 0.0
    no_imp = 0

    history = {
        'train_loss': [], 'val_loss': [],
        'train_acc': [], 'val_acc': []
    }

    for epoch in range(1, epochs + 1):

        # 🔥 Gradual unfreezing
        if epoch == 15:
            print("🔥 Unfreezing EfficientNet...")
            for param in model.effnet.features.parameters():
                param.requires_grad = True

        if epoch == 20:
            print("🔥 Unfreezing ResNet...")
            for param in model.resnet.parameters():
                param.requires_grad = True

        # ===== TRAIN =====
        model.train()
        running_loss = 0.0
        correct = 0
        total = 0

        for xb, yb in train_loader:
            xb = xb.to(device, non_blocking=PIN)
            yb = yb.to(device, non_blocking=PIN)

            optimizer.zero_grad()

            xb, yb_a, yb_b, lam = mixup(xb, yb)

            out = model(xb)

            loss = lam * criterion(out, yb_a) + (1 - lam) * criterion(out, yb_b)

            loss.backward()
            torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
            optimizer.step()

            running_loss += loss.item()
            preds = out.argmax(1)

            correct += (preds == yb).sum().item()
            total += yb.size(0)

        train_loss = running_loss / len(train_loader)
        train_acc = 100.0 * correct / total

        # ===== VALIDATION =====
        model.eval()
        val_loss = 0.0
        correct = 0
        total = 0

        with torch.no_grad():
            for xb, yb in val_loader:
                xb = xb.to(device, non_blocking=PIN)
                yb = yb.to(device, non_blocking=PIN)

                out = model(xb)
                loss = criterion(out, yb)

                val_loss += loss.item()
                preds = out.argmax(1)
                correct += (preds == yb).sum().item()
                total += yb.size(0)

        val_loss /= len(val_loader)
        val_acc = 100.0 * correct / total

        history['train_loss'].append(train_loss)
        history['val_loss'].append(val_loss)
        history['train_acc'].append(train_acc)
        history['val_acc'].append(val_acc)

        scheduler.step()

        print(f'Epoch {epoch}/{epochs} — train_acc {train_acc:.2f}% | val_acc {val_acc:.2f}%')

        # ===== SAVE BEST =====
        if val_acc > best_val:
            best_val = val_acc
            no_imp = 0
            model.eval()

            torch.save({
                'model_state_dict': model.state_dict(),
                'val_acc': best_val,
                'classes': class_names
            }, str(WORK_DIR / 'best_model.pth'))

            print("✅ Saved best model")

        else:
            no_imp += 1
            if no_imp >= patience:
                print("⛔ Early stopping")
                break

    return model, history

# --- Evaluation ---

def evaluate_model(model, loader, classes):
    model.eval(); all_preds=[]; all_labels=[]; all_probs=[]
    with torch.no_grad():
        for xb, yb in loader:
            xb = xb.to(device, non_blocking=PIN)
            out = model(xb)
            prob = torch.softmax(out, dim=1)
            preds = out.argmax(1).cpu().numpy()
            all_preds.extend(preds)
            all_labels.extend(yb.numpy())
            all_probs.extend(prob.cpu().numpy())
    acc = accuracy_score(all_labels, all_preds)
    print('Accuracy:', acc)
    print(classification_report(all_labels, all_preds, target_names=classes, digits=4))
    return np.array(all_labels), np.array(all_preds), np.array(all_probs)

# --- instantiate and train ---
model = HybridCattleClassifier(num_classes=num_classes).to(device)
print('Hybrid model created')

# 🔥 Freeze early layers (reduce overfitting)
# Freeze more of EfficientNet (stronger generalization)
for param in model.effnet.features[:5].parameters():
    param.requires_grad = False

# Freeze early ResNet blocks
for param in model.resnet.layer1.parameters():
    param.requires_grad = False

model, history = train_model(model, train_loader, val_loader, epochs=50, patience=13, lr=2e-4)

# save final
torch.save({'model_state_dict': model.state_dict(), 'classes': class_names}, str(WORK_DIR / 'final_model.pth'))
print('Saved final model to', WORK_DIR / 'final_model.pth')

# evaluate on test set
print('\nEvaluating on test set...')
true, pred, probs = evaluate_model(model, test_loader, class_names)

# generate and save confusion matrix
cm = confusion_matrix(true, pred)
plt.figure(figsize=(10,8))
sns.heatmap(cm, annot=True, fmt='d', xticklabels=class_names, yticklabels=class_names, cmap='Blues')
plt.xlabel('Predicted'); plt.ylabel('True')
plt.tight_layout(); plt.savefig(WORK_DIR / 'confusion_matrix.png', dpi=200); plt.close()
print('Confusion matrix saved to', WORK_DIR / 'confusion_matrix.png')

print('\nAll done. Artifacts saved in', WORK_DIR)
```

```python
# ---- t-SNE FEATURE VISUALIZATION ----
print("Extracting features for t-SNE...")
features = []
labels = []

model.eval()
with torch.no_grad():
    for xb, yb in test_loader:
        xb = xb.to(device)
        feat = model.get_features(xb)
        features.append(feat.cpu().numpy())
        labels.extend(yb.numpy())

features = np.vstack(features)

print("Running t-SNE...")
tsne = TSNE(n_components=2, verbose=1, perplexity=40, n_iter=500)
tsne_results = tsne.fit_transform(features)

plt.figure(figsize=(10,8))
scatter = plt.scatter(tsne_results[:,0], tsne_results[:,1], c=labels, cmap='tab20')
plt.colorbar(scatter, ticks=range(num_classes))
plt.title("t-SNE Embedding of Test Set Features")
plt.savefig(WORK_DIR / "tsne_features.png", dpi=200)
plt.close()

print("Saved t-SNE visualization.")
```

```python
# ---- MISCLASSIFIED IMAGES ----
print("Generating misclassified image grid...")

misclassified = []

model.eval()
with torch.no_grad():
    for xb, yb in test_loader:
        xb = xb.to(device)
        out = model(xb)
        preds = out.argmax(1).cpu()
        yb_cpu = yb.cpu()

        for img, pred_lbl, true_lbl in zip(xb.cpu(), preds, yb_cpu):
            if pred_lbl != true_lbl:
                misclassified.append((img, pred_lbl.item(), true_lbl.item()))

# Show up to 25 images
to_show = misclassified[:25]

fig, axes = plt.subplots(5, 5, figsize=(12,12))
for ax, (img, pred, truth) in zip(axes.flatten(), to_show):
    ax.imshow(denorm(img))
    ax.set_title(f"Pred: {class_names[pred]}\nTrue: {class_names[truth]}", fontsize=8)
    ax.axis('off')

plt.tight_layout()
plt.show()
plt.savefig(WORK_DIR / "misclassified_grid.png", dpi=200)
plt.close()
print("Saved misclassified image grid.")

```

```python
import torch
import torch.nn as nn
from torchvision import transforms
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

# -------------------
# LOAD TRAINED MODEL
# -------------------

checkpoint = torch.load("./working/best_model.pth", map_location="cpu")
class_names = checkpoint["classes"]

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

model = HybridCattleClassifier(num_classes=len(class_names))
model.load_state_dict(checkpoint["model_state_dict"])
model = model.to(device)
model.eval()

print("Model loaded and ready!")


# -------------------
# IMAGE TRANSFORM
# -------------------

inference_transform = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406],
                         std=[0.229, 0.224, 0.225]),
])


# -------------------
# PREDICT FUNCTION
# -------------------

def predict_image(image_path, top_k=5):
    img = Image.open(image_path).convert("RGB")
    tensor = inference_transform(img).unsqueeze(0).to(device)

    with torch.no_grad():
        out = model(tensor)
        prob = torch.softmax(out, dim=1)[0].cpu().numpy()

    top_idx = np.argsort(prob)[-top_k:][::-1]
    top_probs = prob[top_idx]
    top_labels = [class_names[i] for i in top_idx]

    # Display image
    plt.imshow(img)
    plt.axis("off")
    plt.title("Uploaded Image")
    plt.show()

    print("\n🔍 **Prediction Results**")
    print("------------------------------")
    print(f"🎯 BEST PREDICTION: {top_labels[0]} ({top_probs[0]*100:.2f}% confidence)\n")

    print("📌 Top-5 Predictions:")
    for lbl, p in zip(top_labels, top_probs):
        print(f"  - {lbl}: {p*100:.2f}%")

    # Bar plot for top5
    plt.figure(figsize=(8,5))
    plt.barh(top_labels, top_probs)
    plt.gca().invert_yaxis()
    plt.xlabel("Probability")
    plt.title("Top-5 Prediction Confidence")
    plt.show()

    return top_labels[0], top_probs[0]
```

```python

# ---- MISCLASSIFIED IMAGES ----
print("Generating misclassified image grid...")

misclassified = []

model.eval()
with torch.no_grad():
    for xb, yb in test_loader:
        xb = xb.to(device)
        out = model(xb)
        preds = out.argmax(1).cpu()
        yb_cpu = yb.cpu()

        for img, pred_lbl, true_lbl in zip(xb.cpu(), preds, yb_cpu):
            if pred_lbl != true_lbl:
                misclassified.append((img, pred_lbl.item(), true_lbl.item()))

# Show up to 25 images
to_show = misclassified[:25]

fig, axes = plt.subplots(5, 5, figsize=(12,12))
for ax, (img, pred, truth) in zip(axes.flatten(), to_show):
    ax.imshow(denorm(img))
    ax.set_title(f"Pred: {class_names[pred]}\nTrue: {class_names[truth]}", fontsize=8)
    ax.axis('off')

plt.tight_layout()
plt.show()
plt.savefig(WORK_DIR / "misclassified_grid.png", dpi=200)
plt.close()
print("Saved misclassified image grid.")

```

```python
# ---- TOP-K PREDICTION SAMPLE ----
print("Generating Top-5 prediction bar plot...")

sample_img, true_lbl = next(iter(test_loader))
sample_img = sample_img[0].unsqueeze(0).to(device)

with torch.no_grad():
    out = model(sample_img)
    prob = torch.softmax(out, dim=1).cpu().numpy()[0]

top5_idx = np.argsort(prob)[-5:][::-1]
top5_probs = prob[top5_idx]
top5_labels = [class_names[i] for i in top5_idx]

plt.figure(figsize=(8,5))
plt.barh(top5_labels, top5_probs)
plt.gca().invert_yaxis()
plt.title("Top-5 Prediction Probabilities")
plt.xlabel("Probability")
plt.show()
plt.savefig(WORK_DIR / "top5_predictions.png", dpi=200)
plt.close()
```
